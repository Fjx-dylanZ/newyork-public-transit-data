```python
df_subway = npt.retrieve_mta_data_as_df([2020], station=None)
df_subway = npt.turnstile_to_counts(df_subway, granularity='4H', detail_level='station', interpolate=True)

df_citibike = npt.retrieve_citibike_data_as_df(start_date=202001, end_date=202101)
df_citibike = npt.trips_to_counts(df_citibike, granularity='4H', interpolate=True)
#df_citibike = df_citibike.loc[(df_citibike['citibike_station'] == 'Cathedral Pkwy & Broadway') | (df_citibike['citibike_station'] == 'W 106 St & Amsterdam Ave')]
df_citibike = df_citibike.groupby(['citibike_station', 'datetime_rounded']).agg({'start_count': 'sum', 'end_count': 'sum'})
df_citibike = df_citibike.reset_index()
```

```python
df_citibike = npt.retrieve_citibike_data_as_df(start_date=202001, end_date=202101)
df_citibike = npt.trips_to_counts(df_citibike, granularity='4H', interpolate=True)
#df_citibike = df_citibike.loc[(df_citibike['citibike_station'] == 'Cathedral Pkwy & Broadway') | (df_citibike['citibike_station'] == 'W 106 St & Amsterdam Ave')]
df_citibike = df_citibike.groupby(['citibike_station', 'datetime_rounded']).agg({'start_count': 'sum', 'end_count': 'sum'})
df_citibike = df_citibike.reset_index()
```

```python
px.set_mapbox_access_token(open('.mapbox_token').read())
def graph_hospital(hospital_name, hospital_coord, radius=1000, map=False, window='14D'):
    lat = hospital_coord[0]
    lng = hospital_coord[1]
    nearby_subway = npt.find_nearby_stations((lat, lng), radius=radius, kind="subway")
    nearby_citibike = npt.find_nearby_stations((lat, lng), radius=radius/2, kind="citibike")
    nearby_subway['transportation'] = 'subway'
    nearby_citibike['transportation'] = 'citibike'
    nearby = pd.concat([nearby_subway, nearby_citibike], axis=0)
    nearby = pd.concat([
        nearby,
        pd.DataFrame(
            {
                'station': [hospital_name],
                'lat': [lat],
                'lng': [lng],
                'distance': [0],
                'transportation': ['hospital']
            }
        )
    ])
    if (len(nearby_subway) == 0) | (len(nearby_citibike) == 0):
        raise ValueError(f"No subway or citibike stations found within {radius} \n \
                         meters of {hospital_name}. {len(nearby_subway)} subway stations \n \
                         and {len(nearby_citibike)} citibike stations found within {radius/2}")

    if map:
        figure = px.scatter_mapbox(
            nearby,
            lat='lat',
            lon='lng',
            hover_name='station',
            hover_data=['distance'],
            color='transportation',
            color_discrete_sequence=['#1f77b4', '#ff7f0e', '#2ca02c'],
            zoom=13,
            height=500,
            width=800,
            title=f"Subway and Citi Bike Stations within {radius} meters of {hospital_name}",
            mapbox_style='carto-positron'
        )
        figure.update_layout(
            legend=dict(
                orientation='h',
                yanchor='bottom',
                y=1.02,
                xanchor='right',
                x=1
            )
        )
        figure.show()
    else:
        pass

    df_citibike_graph = df_citibike.loc[df_citibike['citibike_station'].isin(nearby_citibike['station'])]
    df_subway_graph = df_subway.loc[df_subway['station'].isin(nearby_subway['station'])]
    max_datetime = df_subway_graph['datetime_rounded'].max()
    min_datetime = df_subway_graph['datetime_rounded'].min()
    df_citibike_graph = df_citibike_graph.loc[(df_citibike_graph['datetime_rounded'] >= min_datetime) & (df_citibike_graph['datetime_rounded'] <= max_datetime)]
    df_graph = (
        df_subway_graph.groupby('datetime_rounded').sum().reset_index()
        .merge(
            df_citibike_graph.groupby('datetime_rounded').sum().reset_index(),
            on='datetime_rounded',
        )
    )
    figure = px.line(
        (
            df_graph
            .resample('1D', on='datetime_rounded').sum().reset_index()
            .assign(
                entries_diff = lambda df: np.log(df['entries_diff'] + 1),
                exits_diff = lambda df: np.log(df['exits_diff'] + 1),
                start_count = lambda df: np.log(df['start_count'] + 1),
                end_count = lambda df: np.log(df['end_count'] + 1)
            )
        ),
        x='datetime_rounded',
        y=['entries_diff', 'exits_diff', 'start_count', 'end_count'],
        labels={
            'datetime_rounded': 'Date',
            'value': 'Count',
            'variable': 'Type'
        },
        title=f"Entries and Exits of Subway and Citi Bike Stations near {hospital_name}",
        color_discrete_sequence=['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e'],
        #symbol_sequence=['circle', 'square', 'circle', 'square'],

    )
    figure.show()

    figure = npt.plot_rolling_pct_change(
        df_graph,
        window=window,
        datetime_col='datetime_rounded'
    )
    figure.update_layout(
        title=f"Rolling Percentage Change of {hospital_name} and Nearby Stations, {window}",
        legend=dict(
            orientation='h',
            yanchor='bottom',
            y=1.02,
            xanchor='right',
            x=1
        )
    )
    figure.show()
            

HOSPITAL_NAME = 'Mount Sinai West'
WINDOW = '14D'
graph_hospital(
    hospital_name=hospital_df.loc[lambda df: df['Name'].str.contains(HOSPITAL_NAME), 'Name'].values[0],
    hospital_coord=hospital_df.loc[lambda df: df['Name'].str.contains(HOSPITAL_NAME), 'coord'].values[0],
    radius=1000,
    map=True,
    window=WINDOW
)
```

```python
def graph_all_hospitals(radius=1000, window="14D", map=False, shrink_factor=2):
    df_nearby_subways = None
    df_nearby_citibikes = None
    df_target_hospitals = None
    for _, row in hospital_df.iterrows():
        nearby_subways = npt.find_nearby_stations(
            (row["lat"], row["lng"]),
            radius=radius,
            kind="subway",
        )
        nearby_citibikes = npt.find_nearby_stations(
            (row["lat"], row["lng"]),
            radius=radius / shrink_factor,
            kind="citibike",
        )
        target_hospital = pd.DataFrame(
            {
                "station": [row["Name"]],
                "lat": [row["lat"]],
                "lng": [row["lng"]],
                "distance": [0],
            }
        )
        if (len(nearby_subways) == 0) or (len(nearby_citibikes) == 0):
            continue
        if df_nearby_subways is None:
            assert df_nearby_citibikes is None and df_target_hospitals is None
            df_nearby_subways = nearby_subways
            df_nearby_citibikes = nearby_citibikes
            df_target_hospitals = pd.DataFrame(target_hospital)
            continue
        df_nearby_subways = pd.concat([df_nearby_subways, nearby_subways])
        df_nearby_citibikes = pd.concat([df_nearby_citibikes, nearby_citibikes])
        df_target_hospitals = pd.concat(
            [
                df_target_hospitals,
                pd.DataFrame(target_hospital),
            ]
        )
    df_nearby_subways = df_nearby_subways.drop_duplicates()
    df_nearby_citibikes = df_nearby_citibikes.drop_duplicates()
    df_nearby_subways['transpotation'] = 'subway'
    df_nearby_citibikes['transpotation'] = 'citibike'
    df_target_hospitals['transpotation'] = 'hospital'

    df_nearby = pd.concat([df_nearby_subways, df_nearby_citibikes, df_target_hospitals])
    df_nearby = df_nearby.reset_index(drop=True)
    
    if map:
        figure = px.scatter_mapbox(
            df_nearby,
            lat="lat",
            lon="lng",
            color="transpotation",
            hover_name="station",
            hover_data=["distance"],
            zoom=12,
            mapbox_style="carto-positron",
            title=f"Nearby Stations within {radius} meters of NYC Hospitals",
        )
        figure.update_layout(
            margin={"r": 0, "t": 0, "l": 0, "b": 0},
            legend=dict(
                orientation="h",
                yanchor="bottom",
                y=1.02,
                xanchor="right",
                x=1,
            ),
        )
        figure.show()

    df_citibike_graph = df_citibike.loc[df_citibike['citibike_station'].isin(df_nearby_citibikes['station'].unique())]
    df_subway_graph = df_subway.loc[df_subway['station'].isin(df_nearby_subways['station'].unique())]
    max_datetime = df_subway_graph['datetime_rounded'].max()
    min_datetime = df_subway_graph['datetime_rounded'].min()
    df_citibike_graph = df_citibike_graph.loc[(df_citibike_graph['datetime_rounded'] >= min_datetime) & (df_citibike_graph['datetime_rounded'] <= max_datetime)]
    df_graph = (
        df_subway_graph.groupby('datetime_rounded').sum().reset_index()
        .merge(
            df_citibike_graph.groupby('datetime_rounded').sum().reset_index(),
            on='datetime_rounded',
        )
    )
    print(df_graph.sort_values('datetime_rounded'))
    figure = px.line(
        (
            df_graph
            #.resample('1D', on='datetime_rounded').sum().reset_index()
            .assign(
                entries_diff = lambda df: np.log(df['entries_diff'] + 1),
                exits_diff = lambda df: np.log(df['exits_diff'] + 1),
                start_count = lambda df: np.log(df['start_count'] + 1),
                end_count = lambda df: np.log(df['end_count'] + 1)
            )
        ),
        x='datetime_rounded',
        y=['entries_diff', 'exits_diff', 'start_count', 'end_count'],
        labels={
            'datetime_rounded': 'Date',
            'value': 'Count',
            'variable': 'Type'
        },
        title=f"Entries and Exits of Subway and Citi Bike Stations near all Hospitals",
        color_discrete_sequence=['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e'],
        #symbol_sequence=['circle', 'square', 'circle', 'square'],

    )
    figure.show()
    figure = npt.plot_rolling_pct_change(
        df_graph,
        window=window,
        datetime_col='datetime_rounded'
    )
    figure.update_layout(
        title=f"Rolling Percentage Change of All hospitals' Nearby Stations, {window}",
        legend=dict(
            orientation='h',
            yanchor='bottom',
            y=1.02,
            xanchor='right',
            x=1
        )
    )
    figure.show()

graph_all_hospitals(map=True, shrink_factor=2, window="4H", radius=700)
```

```python
def graph_all(window: str = "7D"):
    max_datetime = df_subway['datetime_rounded'].max()
    min_datetime = df_subway['datetime_rounded'].min()
    df_graph = (
        df_subway.groupby('datetime_rounded').sum().reset_index()
        .merge(
            (df_citibike.loc[(df_citibike['datetime_rounded'] >= min_datetime) & (df_citibike['datetime_rounded'] <= max_datetime), :]
            .groupby('datetime_rounded').sum().reset_index()),
            on='datetime_rounded'
        )
    )
    figure = px.line(
        (
            df_graph
            #.resample('1D', on='datetime_rounded').sum().reset_index()
            .assign(
                entries_diff = lambda df: np.log(df['entries_diff'] + 1),
                exits_diff = lambda df: np.log(df['exits_diff'] + 1),
                start_count = lambda df: np.log(df['start_count'] + 1),
                end_count = lambda df: np.log(df['end_count'] + 1)
            )
        ),
        x='datetime_rounded',
        y=['entries_diff', 'exits_diff', 'start_count', 'end_count'],
        labels={
            'datetime_rounded': 'Date',
            'value': 'Count',
            'variable': 'Type'
        },
        title=f"Entries and Exits of Subway and Citi Bike Stations",
        color_discrete_sequence=['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e'],
        #symbol_sequence=['circle', 'square', 'circle', 'square'],

    )
    figure.show()
    figure = npt.plot_rolling_pct_change(
        df_graph,
        window=window,
        datetime_col='datetime_rounded'
    )
    figure.update_layout(
        title=f"Rolling Percentage Change of stations, {window}",
        legend=dict(
            orientation='h',
            yanchor='bottom',
            y=1.02,
            xanchor='right',
            x=1
        )
    )
    figure.show()


graph_all(window="14D")
```
